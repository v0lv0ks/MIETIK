```python
y1=e**(x/2)
y2=3+x**(0.5)
sp.plot((y1,(x,-2,10)),(y2,(x,-10,10)))

#ищем корень
def f(x):
    return (3+x**(0.5)-e**(x/2))
x = optimize.brentq(f, 2, 4)
print('x =',x)#значение X
```

```python
from sympy.abc import x

y1=10
y2=x**3-8*x**2+17*x+x**(0.5)
sp.plot((y1,(x,-2,10)),(y2,(x,-10,7)))

#ищем корень
def f(x):
    return (x**3-8*x**2+17*x+x**(0.5)-10)
x1 = optimize.brentq(f, 0, 1)
print('x1 =',x1)#значение X1
x2 = optimize.brentq(f, 2, 3)
print('x2 =',x2)#значение X2
x3 = optimize.brentq(f,4, 5)
print('x3 =',x3)#значение X3
```

```python
from sympy.solvers import nsolve
from sympy import Symbol,sin
from math import *
import numpy as np
import sympy as sp
x = Symbol('x')
eq = e**(3*x)-(3*x)**e
a = nsolve(eq, 0.5)
print ('x = ', a)
print ('f(x) = ', eq.subs(x,a))
print(np.isclose(4.97379915032070e-14, 0))
```

```python
x = Symbol('x')
eq = sp.sin(x)-x/3
a1 = nsolve(eq, -2)
print ('x1 = ', a1)
print ('f(x1) = ', eq.subs(x,a1))

a2 = nsolve(eq, 0)
print ('x2 = ', a2)
print ('f(x2) = ', eq.subs(x,a2))

a3 = nsolve(eq, 2)
print ('x3 = ', a3)
print ('f(x3) = ', eq.subs(x,a3))
```

```python
x = Symbol('x')
y = Symbol('y')
z = Symbol('z')
w = Symbol('w')

f1 = w+x+4*y+3*z-5
f2 = 2*w+3*x+y-2*z-1
f3 = w+2*x+5*y+4*z-3
f4 = w-3*z-9
a1 = nsolve((f1, f2, f3, f4), (x, y, z, w), (-7, 7, 3, 0))
print('x1 = ', a1)
a2 = nsolve((f1, f2, f3, f4), (x, y, z, w), (-0.5, 7, 3, 0))
print('x2 = ', a2)
a3 = nsolve((f1, f2, f3, f4), (x, y, z, w), (0.5, 7, 3, 0))
print('x3 = ', a3)
a4 = nsolve((f1, f2, f3, f4), (x, y, z, w), (13, 7, 3, 0))
print('x4 = ', a4)
```

```python
#C2
from sympy.solvers import nsolve
from sympy import Symbol, sin
x = Symbol('x')
eq = x**2-5*x*sp.sin(3*x)+3
a1= nsolve(eq, -2.7)
print ('x1 = ', a1)
print ('f(x1) = ', eq.subs(x,a1))
a2= nsolve(eq, -2)
print ('x2 = ', a2)
print ('f(x2) = ', eq.subs(x,a2))
a3= nsolve(eq, 2)
print ('x3= ', a3)
print ('f(x3) = ', eq.subs(x,a3))
a4= nsolve(eq, 2.7)
print ('x4 = ', a4)
print ('f(x4) = ', eq.subs(x,a4))
```
