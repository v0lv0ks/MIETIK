```python
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
'''
! Всё в optimize.

Нахождение нулей:
>>> brentq(func, a, b) 
Где (a,b) - интервал, на котором функция меняет знак

>>> fmin(f, x0, full_output=True)
Где аргумент full_output даёт нам точку минимума и значение в ней

'''

'''
Схема исследования функции и построения графика

Исследование функции можно проводить по следующей схеме:
1. Найти область определения функции.
2. Проверить, является ли функция четной, нечетной, периодической.
3. Найти нули функции и промежутки, где значения функции положительны, 
отрицательны.
4. Найти точки разрыва функции. Найти односторонние пределы функции в 
граничных точках области определения и в точках разрыва. Указать
вертикальные асимптоты функции.
5. Найти наклонные асимптоты функции.
6. Вычислить первую производную функции, найти экстремумы и 
промежутки ее возрастания и убывания.
7. Вычислить вторую производную, найти точки перегиба графика, 
промежутки выпуклости вверх и выпуклости вниз.
'''
```

```python
def f(x):
 y = 2**x + x
 return y

x = np.linspace(-2, 2, 100)
plt.plot(x, f(x))
plt.grid(True)

root = optimize.brentq(f, -1, 0)
print(f'Root = {root}')
print(f'f(root) = {f(root)}')

M = optimize.fmin(f, 0, full_output=True)
print(M)
```

```python
# C1
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10,10,1000)
y = np.sin(x) - (x**2)*np.cos(x)
dy_dx = np.gradient(y,x)
d2y_dx2 = np.gradient(dy_dx, x)

zeros = np.where(np.abs(dy_dx) < 0.1)[0]
Z = np.where(np.abs(d2y_dx2) < 0.1)[0]
plt.plot(x,y, label='$y=sin(x) - x^2 cos(x)$')
#plt.plot(x[zeros], y[zeros], 'o', color='red')
plt.plot(x,dy_dx, label="$y'$")
plt.plot(x[Z], dy_dx[Z], 'o', color='green', label="$y''= 0$")
j = 0
for i in x[Z]:
    print(i)
    j += 1
    plt.text(i-1.5, -12 + 3*j + 4*(j%2 * 2 - 1), str(f'{i : 0.2f}'))
plt.legend(bbox_to_anchor=(1.05, 1), loc = 'upper left',
           borderaxespad=0.0, shadow=True)
plt.tight_layout()
plt.grid()
#plt.axis([-10,10, -20,20])
plt.axhline(y=0, color='k')

plt.show()
```
