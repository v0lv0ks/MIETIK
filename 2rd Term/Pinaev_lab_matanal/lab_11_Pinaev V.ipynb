```python
# УПРАЖНЕНИЕ 4
def f(x):
    return (e**(-x))*(np.sin(3*np.pi*x))
def g(x):
    return -f(x)
# [0,2]
x = np.linspace(0,4,1000)
plt.plot(x,f(x))
plt.plot(x,-f(x))
plt.grid()
min = optimize.fmin(f, 0.5, full_output=True, disp=False)
max = optimize.fmin(g, 0, full_output=True, disp=False)

print(f'f(x_min) = {min[1]}')
print(f'f(x_max) = {f(max[0][0])}')

f_x_min =min[1]
f_x_max = f(max[0][0])
f = open('t.txt','a')

f.write("| {x:7f} | {y:11.8f} |\n".format(x=f_x_min,y=f_x_max))
f.close()
```
