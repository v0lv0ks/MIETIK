```python
import math
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.1)
plt.figure(figsize = (10, 10))
y = [(2 ** i) / math.gamma(i + 1) for i in x]  
plt.plot(x,y)
plt.xlabel('Ось x') 
plt.ylabel('Ось y')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.show ()
```
