```python
import matplotlib.pyplot as plt

import numpy as np
x=np.array([1+3j,3+4j,x[0]+x[1],x[0]-x[1]])
orig=np.zeros(4)
plt.plot()
q=plt.quiver(orig,orig,np.real(x),np.imag(x),
             angles='xy', scale_units='xy',scale=1,color="green")
plt.axis([-9,9,-9,9])

plt.axhline(y=0,color='k')
plt.axvline(x=0,color='k')
plt.grid(True)
plt.title('Copmlex numbers visualised by quivers')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.show()
# Геометрический смысл суммы комплексных чисел состоит в том, что мы берём вектор, изображающий
# первое число, и прибавляем (по правилу треугольника, например) к этому вектору вектор, 
# изображающий второе число.
# Аналогично с разностью, только мы не прибавляем вектор, а вычитаем.
```

```python
    
z=2*(3**0.5)-2j
r=np.abs(z)#ищем длину вектора Z
phi=np.angle(z)#ишем угол
k=np.arange(1,5)
zroot=r**(1/4)*(np.cos((phi+2.0*np.pi*k)/4)+1j*np.sin((phi+2.0*np.pi*k)/4))#ищем корни уравнений по формуле Муавра
print(zroot)
plt.title('$\sqrt[4]{2\sqrt{3}-2i}$ complex roots')#легенда
plt.plot(np.real(zroot), np.imag(zroot), '*g')#
zroot=np.append(zroot,zroot[0])#продублировали первый элемент в массиве корней(zroot) 
plt.plot(np.real(zroot),np.imag(zroot),'*-.')#соединяем корни отрезками


plt.xlabel('Ось реальных чисел')
plt.ylabel('Ось мнимых чисел')
plt.axhline(y=0,color='k')
plt.axvline(x=0,color='k')
plt.grid(True)
plt.axis('equal')
t=np.linspace(0,2*np.pi,100)#
plt.plot(r**(1/4)*np.cos(t),r**(1/4)*np.sin(t),':')#
```

```python
x = np.linspace(-1,1,40)
arg1, = plt.plot(x,x*np.tan( np.pi/4 ), 'r-.', label = 'Arg(z) = $pi/4$')
arg2, = plt.plot(x,x*np.tan(5*np.pi/6), 'g:', label = 'Arg(z) = $5pi/6$')
plt.legend(handles = [arg1, arg2])

plt.grid(True)
plt.axis('equal')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.title('Множество точек Arg(z) = const')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.show()
```
