```python
import sympy as sp
from math import *
from sympy.abc import v
import numpy as np
u=0.002
T1=100
T2=300
N=6.022*10**23
k=1.381*10**(-23)
m=u/N
y1=v**2*4*np.pi*((m/(2*np.pi*k*T1))**(1.5))*e**((-m*v**2)/(2*k*T1))
y2=v**2*4*np.pi*((m/(2*np.pi*k*T2))**(1.5))*e**((-m*v**2)/(2*k*T2)) 
sp.plot((y1,(v,0,4000)),(y2,(v,0,5000)))
```
