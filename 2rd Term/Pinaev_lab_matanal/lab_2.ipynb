```python
import numpy as np
x=np.linspace(-2.0*np.pi,2.0*np.pi,100)
plt.figure(figsize=(10,10)) #изменение размера
h1, = plt.plot(x,np.sin(x),'pr',label='y=sin(x)-2')
h2, = plt.plot(x,np.sin(x)-2,'xb',label='y=sin(x)-2')
h3, = plt.plot(x,np.sin(x)+1,'om:',label='y=sin(x)+1')

plt.grid(True)

plt.axhline(y=0, color='k') # отобразим ось x
plt.axvline(x=0, color='k') # отобразим ось y

plt.xlabel('x') #подпишем ось х
plt.ylabel('y') #подпишем ось у

plt.show()
```

```python
x=np.linspace(-2.0*np.pi,2.0*np.pi,100)
h1, = plt.plot(x,np.cos(x),'pr',label='y=cos(x)')
h2, = plt.plot(x,2*np.cos(x),'xb',label='y=2cos(x)')
h3, = plt.plot(x,0.3*np.cos(x),'om:',label='y=0.3cos(x)')
h4,= plt.plot(x,-np.cos(x),'og:',label='y=-cos(x)')
plt.grid(True) # добавляем линии сетки на график
plt.legend(handles=[h1,h2,h3,h4]) # выводим в графическом окне область с легендой-справкой о названиях графиков
plt.axhline(y=0, color='k') # отобразим ось x
plt.axvline(x=0, color='k') # отобразим ось y
plt.xlabel('x') #подпишем ось х
plt.ylabel('y') #подпишем ось у
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,10)) #изменение размера

def f(x): #определение функции f(X)
    return np.abs(np.abs(x) - 2)

x=np.linspace(-5.0, 5.0 ,100) #промежуток
# График f(x)
plt.subplot(2,3,1)
plt.plot(x, f(x))
plt.grid(True) 
# График f(x - 2), сдвинут влево на 2 относительно f(x)
plt.subplot(2,3,2)
plt.plot(x,f(x - 2))
plt.grid(True)
# График f(x + 2), сдвинут вправо на 2 относительно f(x)
plt.subplot(2,3,3)
plt.plot(x,f(x + 2))
plt.grid(True)
# График f(2x) шире в 2 раза, чем f(x)
plt.subplot(2,3,4)
plt.plot(x,f(2 * x))
plt.grid(True)
# График f(0.5x) уже в 2 раза, чем f(x)
plt.subplot(2,3,5)
plt.plot(x,f(0.5 * x))
plt.grid(True)
# График f(-x) не отличается от f(X), так как x находится под модулем
plt.subplot(2,3,6)
plt.plot(x,f(-x))
plt.grid(True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-3,4,1) # массив степеней 
z = np.ones(7)*10 # массив вида [10 10 ... 10]
y = np.power(z,x) # поэлементное возведение в степень
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.semilogy(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,10)) #изменение размера
x=np.linspace(0, np.pi,100)
handle1, = plt.plot(x,np.cos(x), ':g',label='y = cos(x)')
handle2, = plt.plot(np.cos(x), x, '-.r',label='cos(y) = x')
handle3, = plt.plot(x, x, '-b',label='y = x')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.legend(handles=[handle1,handle2,handle3])
plt.grid()
plt.axhline(y=0, color='k', label = 'Y') # отобразим ось x
plt.axvline(x=0, color='k', label = 'X') # отобразим ось y
plt.title('График y = cos(x) и график, обратый этой функции')
plt.axis('equal')
plt.show()
```
