```python
#1 предел
n = sp.Symbol('n')
xn = sp.Abs(sp.sqrt((n**2+2*n+3))-(sp.sqrt((n**2-2*n-1))))
e = 0.01
a = sp.limit(xn, n, sp.oo)#ищем предел равный a
e1 = sp.solve_univariate_inequality((xn-a) < e, n, relational=False)#расписываем предел по определению и решаем модульное неравенство
e2 = sp.solve_univariate_inequality(n > 0, n, relational=False)#убираем n<0, т.к n-натур.числа
sp.pprint(sp.Intersection(e1,e2))#ищем пересечение
print('n(0.01) = 200 (При n>2000 все значения все значения функции будут лежать в коридора a-0.01<a<a+0.01)')

#2 предел
n = sp.Symbol('n')
xn = sp.Abs(sp.sqrt((n**2+2*n+3))-(sp.sqrt((n**2-2*n-1))))
e = 0.001
a = sp.limit(xn, n, sp.oo)#ищем предел равный a
e1 = sp.solve_univariate_inequality((xn-a) < e, n, relational=False)#расписываем предел по определению и решаем модульное неравенство
e2 = sp.solve_univariate_inequality(n > 0, n, relational=False)#убираем n<0, т.к n-натур.числа
sp.pprint(sp.Intersection(e1,e2))#ищем пересечение
print('n(0.001) = 2000 (При n>2000 все значения все значения функции будут лежать в коридора a-0.01<a<a+0.01)')


```
