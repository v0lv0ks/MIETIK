1 упражнение

```python
import sympy as sp
import numpy as np
from sympy import *
from scipy.interpolate import approximate_taylor_polynomial 
import matplotlib.pyplot as plt
```

```python

x=sp.Symbol('x')
a=sp.Symbol('a')
t1=sp.limit(-sp.cos(np.pi*x/2)+a,x,1)
t2=sp.limit(np.pi-sp.pi*x/2,x,1)
print(t1,'=',t2)

```

```python

x=sp.Symbol('x')
f1=-sp.cos(sp.pi*x/2)+a
f2=sp.pi-sp.pi*x/2
y1=diff(f1,x,1)
y2=diff(f2,x,1)
print(y1.subs(x,1))
print(y2)
```
