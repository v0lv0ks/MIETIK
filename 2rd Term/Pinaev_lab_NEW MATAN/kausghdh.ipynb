```python
import sympy as sp
from sympy import integrate, symbols, diff
```

```python
x,y,t,z = symbols("x y t z")
alpha, betta = symbols("a b")
x = integrate(sp.cos(z)/z, (z, 1, t))
y = integrate(sp.sin(z)/z, (z, 1, t))
A = integrate(sp.sqrt( diff(x,t)**2 + diff(y,t)**2 ), (t, 0, sp.pi) )
sp.pprint(A)
print(A.evalf())
sp.pprint(diff(x,t))
sp.pprint(diff(y,t))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as negr
```

```python
t = np.linspace(-2*np.pi, 2*np.pi)
negr.simpson(1/t, t)
x = np.cos(t) / t
y = np.sin(t) / t
plt.figure(figsize=(20,14))
plt.plot(x,y)
plt.grid()
```
