```python
import matplotlib
from matplotlib.pyplot import *
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt



dp1 = 0.08
dp = 0.06
dt = 0.01
fig  = plt.figure()
ax = fig.add_subplot(111)

lnp = [ 1.74, 1.97, 2.13, 2.35, 2.46  ] 
lnt = [ 7.63, 7.675, 7.69, 7.74, 7.76 ]
plt.grid()
print(list(zip(lnp,lnt)))
for x,y in zip(lnp,lnt):
    ax.plot(y,x,'ro')
    if  ( x == 2.35 and y == 7.74 ) or (x == 2.46 and y == 7.76):
        rect = matplotlib.patches.Rectangle((y - dt,x - 0.06), 2 * dt, 2 * 0.06, fill = '')
        ax.add_patch(rect)
    else:
        rect = matplotlib.patches.Rectangle((y - dt,x - dp1), 2 * dt, 2 * dp1, fill = '')
        ax.add_patch(rect)

h = np.arange(7.6, 7.8, 0.01)
def f(h):
    return 3.86*h - 27.6

ax.plot(lnt, lnp, color = 'red')
"""plt.plot([7.76,7.63], [2.46,1.74], color = "orange", )"""

plt.plot([ lnt[0] + dt, lnt[-1] - dt], [lnp[0] - dp1,lnp[-1] + dp], color = "green")
plt.plot([ lnt[0] - dt, lnt[-1] + dt], [lnp[0] + dp1,lnp[-1] - dp], color = "green")

N1 = 1 / (  ( lnt[-1] - dt - ( lnt[0] + dt) ) / ( lnp[-1] + dp - (lnp[0] - dp1 ) ) )
N2 = 1 / (  ( lnt[-1] + dt - (lnt[0] - dt ) ) / ( lnp[-1] - dp - (lnp[0] + dp1 ) ) )

print(lnt[-1] - dt, ( lnt[0] + dt), ( lnp[-1] + dp),  (lnp[0] - dp1 ) )

print(lnt[-1] + dt, ( lnt[0] - dt), ( lnp[-1] - dp),  (lnp[0] + dp1 ) )

plt.plot(h,f(h))
print("N1 =",N1)

print("N2 =",N2 )

print("<N> =", (N1 + N2) / 2)

print("delta (<N>) =", (N1 - N2) / 2)

print("Ответ: N = (5.8 +- 1.98)")
```
