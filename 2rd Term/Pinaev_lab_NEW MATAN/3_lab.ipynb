```python

def sum_of_series(Expression, epsilon):
    n = sp.Symbol(' n ')
    i = 0
    a = sp.limit( ( Expression.subs(n, n + 1 )) / (Expression.subs(n,n )), n, sp.oo)
    print(a)
    while (True):
        i += 1
        q = max( (Expression.subs(n, i + 1) / Expression.subs(n, i)), a)
        if (q < 1):
            R_k = Expression.subs(n, i) / (1 - q)
            if (R_k <= epsilon):
                break
    sum = 0
    for j in range(1, i + 1):
        sum += Expression.subs(n, j)
    return f"Sum [{1} -> {i}] = {sum.evalf()}\nR_k <= {R_k.evalf()} \n\nn_0 = {i} \nFor all k >= {i} Sum of a_k <= epsilon = {epsilon} \n------------\n"

n = sp.Symbol(' n ')
print(sum_of_series(  n / ( 2**(n-1)) , epsilon = 0.001))
# Начиная с k = n_0 сумма остатка < epsilon
```

```python
def sum_of_seriess(Expression, epsilon):
    n = sp.Symbol(' n ')
    i = 0
    while (True):
        i += 1
        R_k = abs( Expression.subs(n, i + 1) )
        if (R_k <= epsilon):
            break
    sum = 0
    for j in range(1, i + 1):
        sum += Expression.subs(n, j)
    return f"Sum [{1} -> {i}] = {sum.evalf()}\nR_k <= {R_k.evalf()} \n\nn_0 = {i} \nFor all k >= {i} Sum of a_k <= epsilon = {epsilon}\n------------\n"
```

```python
n = sp.Symbol(' n ')
a = sp.limit(( (-1)**(n - 1) ) / (2 * ( n  + 1) )**n, n, sp.oo)
print("lim(an) = ", a)
print( sum_of_seriess( ( (-1)**(n - 1) ) / (2 * ( n  + 1) )**n, epsilon = 0.001 ) )
```
