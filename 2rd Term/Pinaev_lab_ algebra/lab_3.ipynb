```python
a = np.array([5, 4])
b = np.array([-1,5])
print ('Вектор a:', a)
print ('Вектор b:', b)
cos_angle = np.dot(a, b) / np.linalg.norm(a) / np.linalg.norm(b)
print ('Косинус угла между a и b:', cos_angle)
alfa= np.arccos(cos_angle) # арккосинус угла
alfa1=math.degrees(alfa)
print ('Угол в градусах:', alfa1)
```

```python
b = np.array([1, -5])
c = np.array([6,-1])
print ('Вектор b:', b)
print ('Вектор c:', c)
cos_angle = np.dot(b, c) / np.linalg.norm(b) / np.linalg.norm(c)
print ('Косинус угла между b и c:', cos_angle)
alfa= np.arccos(cos_angle) # арккосинус угла
alfa2=math.degrees(alfa)
print ('Угол в градусах:', alfa2)
```

```python
OA = np.array([1,3,-1])
OB = np.array([2,-1,4])
OC = np.array([5,0,3])

a=OB-OA
b=OC-OB
c=OC-OA

print('a:', a)
print('b:', b)
print('c:', c)

x1=np.sqrt(sum(a**2))
x2=np.sqrt(sum(b**2))
x3=np.sqrt(sum(c**2))

p=(x1+x2+x3)/2
s=sqrt(p*((p-np.linalg.norm(a))*(p-np.linalg.norm(b))*(p-np.linalg.norm(c))))
print('S(ABC)=',s)

x=np.linalg.norm(np.cross(a,c))
x1=s/x
print('S(ABC)/|AB,AC| =',x1)

```
