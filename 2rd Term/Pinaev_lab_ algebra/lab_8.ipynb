```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from sympy import *
```

1 ex

```python

x=np.linspace(-1,1,200)
y=np.linspace(-1,1,200)
[x,y]=np.meshgrid(x,y)
z=lambda x,y: x**2-y**2
Z=z(x,y)
fig=plt.figure(figsize=(10,10))
ax=plt.axes(projection='3d')
color=plt.get_cmap('gnuplot')

surf = ax.plot_surface(x,y,Z,rstride = 8,cstride = 8,alpha = 0.8,cmap = color)
cset = ax.contourf(x,y,Z,zdir ='z',offset = np.min(Z),cmap=color)
cset = ax.contourf(x,y,Z,zdir ='x',offset =-2,cmap = color)
cset = ax.contourf(x,y,Z,zdir ='y',offset = 2,cmap = color)

ax.set_xlabel('X-axis')
ax.set_xlim(-2, 2)
ax.set_ylabel('Y-axis')
ax.set_ylim(-2, 2)
ax.set_zlabel('Z-axis')
ax.set_zlim(np.min(Z), np.max(Z))
ax.set_title('График гиперболического параболоида $x^2-y^2=z$ с проекциями')
#ax.text(-2,-2,2,'Сечение OYZ: $z=x^2+c$|Сечение OXZ: $z=-y^2+c$ |Сечение OXY: $x^2-y^2=c$')
plt.show()
```

2 ex

```python
fig = plt.figure(figsize =(10, 10))
ax = plt.axes(projection ='3d')
#Коэффициенты уравнения
R=1
u = np.linspace(0, 2*np.pi, 200) # Угол на плоскости
v = np.linspace(-np.pi/2, np.pi/2, 200) # Угол в пространстве
#Уравнение эллипсоида
x = R* np.outer(np.cos(u), np.cos(v))
y = R* np.outer(np.sin(u), np.cos(v))
z = R * np.outer(np.ones_like(u), np.sin(v))
my_cmap = plt.get_cmap('magma')
surf = ax.plot_surface(x,y,z,rstride = 10,cstride = 10,alpha = 0.6,cmap = my_cmap)
cset = ax.contourf(x,y,z,zdir ='z',offset = -2,cmap = my_cmap)
cset = ax.contourf(x,y,z,zdir ='x',offset =-2,cmap = my_cmap)
cset = ax.contourf(x,y,z,zdir ='y',offset = 2,cmap = my_cmap)
ax.set_xlabel('X-axis'); ax.set_xlim(-2, 2)
ax.set_ylabel('Y-axis'); ax.set_ylim(-2, 2)
ax.set_zlabel('Z-axis'); ax.set_zlim(-2, 2)
ax.set_title('График сферы  $x^2+y^2+z^2=1$ с проекциями')
plt.show()

#np.ones_like -- Возвращает массив единиц той же формы и типа, что и данный массив
#np.outer() -- это поэлементное умножение одного массива на  другой
```

```python
fig = plt.figure(figsize =(12, 12))
ax = plt.axes(projection ='3d')
#Коэффициенты уравнения
coefs = (1, 2,3)
rx, ry, rz = np.sqrt(coefs)
u = np.linspace(0, 2*np.pi, 200) # Угол на плоскости
v = np.linspace(-2,2, 200) # Угол в пространстве
#Уравнение эллипсоида
x = rx * np.outer(np.cos(u), np.cosh(v))
y = ry * np.outer(np.sin(u), np.cosh(v))
z = rz * np.outer(np.ones_like(u), np.sinh(v))
my_cmap = plt.get_cmap('rainbow')
surf = ax.plot_surface(x,y,z,rstride = 10,cstride = 10,alpha = 0.6,cmap = my_cmap)
cset = ax.contourf(x,y,z,zdir ='z',offset = -10,cmap = my_cmap)
cset = ax.contourf(x,y,z,zdir ='x',offset =-10,cmap = my_cmap)
cset = ax.contourf(x,y,z,zdir ='y',offset = 10,cmap = my_cmap)
ax.set_xlabel('X-axis'); ax.set_xlim(-10, 10)
ax.set_ylabel('Y-axis'); ax.set_ylim(-10, 10)
ax.set_zlabel('Z-axis'); ax.set_zlim(-10, 10)
ax.set_title('График однополосного гиперболоида с  проекциями|OYZ: $y^2/b^2-z^2/c^2=1$|OXZ: $x^2/b^2-z^2/c^2=1$ |OXY: $x^2/a^2+y^2/b^2=1$')
```

```python
# уравнение прямой, проходящей через две точки р1 и р2
p1, p2 = Point(8,-1,-3), Point(-4,3,3) 
L = Line(p1, p2) 
L=L.arbitrary_point()
print('Параметрические уравнения прямой:',L)
x,z,t=symbols('x,y,t')
P=x**2-2*x-2*z
PP=P.subs([(x,L[0]),(z,L[1])])
print(PP) # уравнение относительно параметра t
X=solve(PP)
print('Значения параметра t=',X)

t=X[0]
M1=[8 - 12*t, 4*t - 1, 6*t - 3]
t=X[1]
M2=[8 - 12*t, 4*t - 1, 6*t - 3]
print('M1 =',M1)
print('M2 =',M2)

ax1 = plt.axes(projection ='3d')

x = np.linspace(-5, 7, 20)
y = np.linspace(-3, 4, 20)
x, y = np.meshgrid(x, y)
z =(x**2-2*x)/2
ax1.plot_wireframe(x, y, z,color='b')
ax1.plot_wireframe(x, y, z, color='b')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.set_zlabel('Z-axis')


x1 = np.array([8-12*X[0],8-12*X[1]])
y1 = np.array([4*X[0] - 1, 4*X[1] - 1])
z1 = np.array([6*X[0] - 3, 6*X[1] - 3])
ax1.plot(x1, y1, z1,color='r')
ax1.plot(M1[0], M1[1], M1[2],'or',markersize=10,label='точка M1')
ax1.plot(M2[0], M2[1], M2[2],'ok',markersize=10,label='точка M2')
#ax1.legend()
ax1.view_init(90, 0)
plt.show()
```
