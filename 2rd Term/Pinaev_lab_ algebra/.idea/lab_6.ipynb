```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
import sympy as sp
from sympy.abc import alpha
```

1 ЗАДАНИЕ

```python

x1, y1 = 2*np.sqrt(3), -np.sqrt(3)
x0, y0 = -1, -2
a=-1
b=-2
alpha1 = (np.pi)/ 6
def NEW_COORD(x_1,y_1, angle,a,b):
    x = (x_1) * np.cos(angle) - y_1 * np.sin(angle) + a
    y = (y_1) * np.cos(angle) + x_1 * np.sin(angle) + b
    return (x,y)
def OLD_COORD(x,y, angle,a,b):
    x_1 = (x - a) * np.cos(-angle) - (y - b) * np.sin(-angle) 
    y_1 = (y - b) * np.cos(-angle) + (x - a) * np.sin(-angle) 
    return (x_1,y_1)

x, y = NEW_COORD(x1,y1, alpha1,a,b)
print(x,y)
x_1, y_1 = OLD_COORD(x,y, alpha1,a,b)
print(x_1,y_1)
plt.arrow(-1,0,6,0, width=0.05, color='black')
plt.arrow(0,-2,0,7, width=0.05, color='black')

x3,y3 = np.cos(-alpha1), -np.sin(-alpha1)
x4,y4 = np.sin(-alpha1),np.cos(-alpha1)

plt.arrow(x0,y0,x3,y3, width=0.05, color='blue')
plt.arrow(x0,y0,x4,y4, width=0.05, color='blue')

plt.plot(x,y, 'or')#новая точка
plt.plot(x1,y1, 'or')#старая точка
plt.text(x - 0.5, y, 'A`')
plt.text(x1 + 0.2, y1, 'A')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.grid()
#print(x,y)
print(x1,y1)
```

2 ЗАДАНИЕ

```python
# создаём символьные переменные
a=3
b=2
x1,y1,alfa = symbols('x1 y1 alfa')
x = x1 * cos(alfa) - y1 * sin(alfa)
y = y1 * cos(alfa) + x1 * sin(alfa)
f = y - 2 * x+4
# решаем систему уравнений
y1 = solve(f, 'y1')
# узнаем уравнение для дальнейших вычислений
print('Решение системы:', y1)
y1 = ((-x1*sin(alfa) + 2*x1*cos(alfa) - 4)/(2*sin(alfa) + cos(alfa)))
yy = y1.subs(alfa, -pi / 4)
print('Уравнение повернутой прямой:', yy)
# строим прямые
x = np.linspace(-1, 5, 100)
plt.plot(x+3, 2+(-sqrt(2)*(3*sqrt(2)*x/2 - 4)))#новая прямая
plt.plot(x, 2 * x-4)#старая прямая
# обрабатываем точку
alfa = -np.pi/4
x0 = 1; y0 = -2
x2 = x0 * np.cos(-alfa) - y0 * np.sin(-alfa)+a
y2 = y0 * np.cos(-alfa) + x0 * np.sin(-alfa)+b
print('Координаты точки в новой системе:', x2, y2)
# координатные оси
plt.arrow(-2, 0, 6, 0, width=0.05, color='black')
plt.arrow(0, -2, 0, 6, width=0.05, color='black')
# координаты нового базиса
x3 = np.cos(alfa); y3 = -np.sin(alfa)
x4 = np.sin(alfa); y4 = np.cos(alfa)
plt.arrow(3, 2, x3, y3, width=0.05, color='blue')
plt.arrow(3, 2, x4, y4, width=0.05, color='blue')
# строим точки
plt.plot(x0, y0, 'or',)
plt.plot(x2, y2, 'or')
plt.axis('equal')
plt.text(x0+0.5, y0,'M')
plt.text(x2+0.5, y2,'M`')
plt.grid()
```
