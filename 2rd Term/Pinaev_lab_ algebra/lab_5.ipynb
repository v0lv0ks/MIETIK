```python
print('Направляющий вектор данной прямой:',(2,-4,1))
M0=Point(-1,2,0)
M=Point(2,-1,-3)
P1=Line(M0,direction_ratio=(2,-4,1))
P2=Line(M,direction_ratio=(2,-4,1))
M0M=Line(M0,M)

#Так как прямые должны быть параллельны, следовательно
# В качестве направляющего вектора можно использовать один и тот же
print('Получаем новое уравнение:',P1.equation())
print('Сделаем его параметрическим:',P2.arbitrary_point())
print('Найдем прямую M0M,лежащую в плоскости наших P1 и P2:',M0M.equation())
print('Найдем параметрическое ур.M0M:',M0M.arbitrary_point())
print('Найдем норм.вектор желаемой плоскости:', np.cross([3,-3,-3],[2,-4,1]))
L=Plane(M,(-15,-9,-6))
print('Напишем ее уравнение:',L.equation()/3)
print('Проверим ортогональность:',np.dot([-15,-9,-6],[2,-4,1]))

A1 = -5; B1 = -3; C1 = -2; D1 = 1
f1 = lambda x, y: (-A1 * x - B1 * y- D1) / C1
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(1, 1, 1, projection='3d')
xval = np.linspace(-4, 4, 100)
yval = np.linspace(-4, 4, 100)
x, y = np.meshgrid(xval, yval)
z1 = f1(x, y)
surf = ax.plot_surface(x, y, z1, rstride=10, cstride=10, cmap=cm.plasma).set_alpha(0.3)
p1 = np.array([-1,2,0]) # точка на пл-ти
v1 = np.array([2,-4,1]) # направляющий вектор
p2 = np.array([2,-1,-3]) # точка на пл-ти
v2 = np.array([2,-4,1]) # направляющий вектор
p3 = np.array([-1,0,3]) # точка на пл-ти
v3 = np.array([5,3,2]) # нормальный вектор

ax.quiver(*p1, *(v1/2), color='Red') # строим вектор q1
ax.quiver(*p2, *(v2/2), color='Red') # строим вектор q2
ax.quiver(*p3, *v3, color='Blue') # строим вектор n

ax.plot([-1, 1], [2,-2], [0, 1],'--',color='Blue')
ax.plot([-1, -3], [2, 6], [0, -1],'--',color='Blue')
ax.plot([2, 4], [-1,-5], [-3, -2],'--',color='Blue')
ax.plot([2, 0], [-1, 3], [-3, -4],'--',color='Blue')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def F(A,A1,vector1,vector2):
    L1=Line(A,direction_ratio=vector1)#задаем прямую L1
    L2=Line(A1,direction_ratio=vector2)#задаем прямую L2
    P=Plane(A,normal_vector=vector1)#строим пл-ть перпендикул-ую L1 
    d=P.projection_line(L2)#ищем проекцию L2 на пл-ть P
    return A.distance(d)#находим расстояние

print('Расстояние между прямыми:',F(Point(2,-1,0),Point(7,1,3),[3,4,2],[3,4,2]))    
```
